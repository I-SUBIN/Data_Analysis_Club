```python
# 필요한 패키지 임포트
import numpy as np
import pandas as pd
import mglearn
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
```

# 데이터 불러오기

```python
#데이터 불러오기
iris=pd.read_csv('iris.csv')
iris
```

```python
#데이터 자르기
X = iris[['Petal.Length','Petal.Width']]
Y = iris[['Species']]
```

```python
print(X)
```

# K-means clustering

```python
kmeans_4 = KMeans(n_clusters=4).fit(X)  # 클러스터를 4개라 할 때
print(kmeans_4.labels_)  # (0, 1, 2, 3) 4개 종류로 분류됨

kmeans_3 = KMeans(n_clusters=3).fit(X)  # 클러스터를 3개라 할 때
print(kmeans_3.labels_)  # (0, 1, 2) 3개 종류로 분류됨

kmeans_2 = KMeans(n_clusters=2).fit(X)  # 클러스터를 2개라 할 때
print(kmeans_2.labels_)  # (0, 1) 2개 종류로 분류됨
```

# 시각화

```python
# 군집=2 일때
mglearn.discrete_scatter(X['Petal.Length'],X['Petal.Width'],kmeans_2.labels_, markers='o')
```

```python
# 군집=3 일 때
mglearn.discrete_scatter(X['Petal.Length'],X['Petal.Width'],kmeans_3.labels_, markers='o')  
```

```python
# 군집=4 일 때
mglearn.discrete_scatter(X['Petal.Length'],X['Petal.Width'],kmeans_4.labels_, markers='o')  
```

# agglomerative clustering

```python
# 군집분석 수행
agg_2 = AgglomerativeClustering(n_clusters=2).fit(X)  # 클러스터를 2개라 할 때
print(agg_2.labels_)  # (0, 1) 2개 종류로 분류됨

agg_3 = AgglomerativeClustering(n_clusters=3).fit(X)  # 클러스터를 3개라 할 때
print(agg_3.labels_)  # (0, 1, 2) 3개 종류로 분류됨

agg_4 = AgglomerativeClustering(n_clusters=4).fit(X)  # 클러스터를 4개라 할 때
print(agg_4.labels_)  # (0, 1, 2, 3) 2개 종류로 분류됨
```

```python
# 군집=2 일때
mglearn.discrete_scatter(X['Petal.Length'],X['Petal.Width'],agg_2.labels_, markers='^')
```

```python
# 군집=3 일때
mglearn.discrete_scatter(X['Petal.Length'],X['Petal.Width'],agg_3.labels_, markers='^')
```

```python
# 군집=4 일때
mglearn.discrete_scatter(X['Petal.Length'],X['Petal.Width'],agg_4.labels_, markers='^')
```

# DBSCAN clustering

```python
# 군집분석 수행
dbscan_1 = DBSCAN(eps=0.5, min_samples=5).fit(X)  # 기본값
print(dbscan_1.labels_)
dbscan_2 = DBSCAN(eps=1, min_samples=5).fit(X)  # eps = 1일 때
print(dbscan_2.labels_)
dbscan_3 = DBSCAN(eps=1.5, min_samples=5).fit(X)  # eps = 1.5일 때
print(dbscan_3.labels_)
dbscan_4 = DBSCAN(eps=2, min_samples=5).fit(X)  # eps = 2일 때
print(dbscan_4.labels_)
```

```python
# 군집=4 일때
mglearn.discrete_scatter(X['Petal.Length'],X['Petal.Width'],dbscan_1.labels_, markers='^')
```
